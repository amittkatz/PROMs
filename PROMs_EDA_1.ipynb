#Group by the patient data

```python
def group_and_combine(df):
    # Group by 'Not_ID' and 'Actual date of entry into the operating room'
    grouped = df.groupby(['Not_ID', 'Actual date of entry into the operating room']).agg(lambda x: ', '.join(set(filter(lambda v: v and v.lower() != "nan", map(str, x)))) if any(x.notna()) else None)

    # Reset the index
    grouped.reset_index(inplace=True)

    return grouped
```

```python
# Call the function to group and combine the rows
df_2019 = group_and_combine(df_2019)
df_2020 = group_and_combine(df_2020)
df_2021 = group_and_combine(df_2021)
```

```python
#In some of the data, there was probably an error in typing or additional measurement, so if more than one data is written, we will average them.

def fix_multiple_values(df):
    for column in ['height', 'Weight']:
        # Check if the values in the column are strings
        if df[column].dtype == 'O':
            # Split the values by comma and convert to float if possible
            df[column] = df[column].apply(lambda x: [float(val.strip()) if val and val.strip().replace('.', '', 1).isdigit() else val.strip() for val in str(x).split(',')] if isinstance(x, str) else x)

            # Calculate the average for rows with multiple numeric values
            df[column] = df[column].apply(lambda x: (sum([v for v in x if isinstance(v, float)]) / len([v for v in x if isinstance(v, float)])) if isinstance(x, list) and len([v for v in x if isinstance(v, float)]) > 1 else x[0] if isinstance(x, list) and len([v for v in x if isinstance(v, float)]) == 1 else x)
```

```python
#apply the function on the df
fix_multiple_values(df_2019)
fix_multiple_values(df_2020)
fix_multiple_values(df_2021)
```

##checking for null values

```python
#create a file with all the null values
dataframes = [df_2019, df_2020, df_2021, df_proms]
years = ['2019', '2020', '2021', 'PROMs']

missing_summary = []

# Calculate the count and percentage of missing values
for df, year in zip(dataframes, years):
    missing_count = round(df.isnull().sum())
    missing_percentage = round((missing_count / len(df)) * 100, 3)
    missing_summary.append(pd.concat([missing_count, missing_percentage], axis=1, keys=[f"{year} Missing Count", f"{year} Missing %"]))

# Create a summary DataFrame
missing_summary_df = pd.concat(missing_summary, axis=1)

# Save the summary DataFrame to a CSV file
missing_summary_df.to_csv("missing_summary.csv")
```

# time of surgery

creating a new variable that conatin the surgey time, we will use 'cut time' becuase it has the smallest amount of missing values and it will be the most accurate time for the surgery itself

```python
import pandas as pd

def time_of_surgery(df):
    """
    Calculate surgery duration in minutes for a given DataFrame.

    This function takes a DataFrame as input, calculates the surgery duration in minutes
    for each surgery record in the DataFrame, and updates the DataFrame in place.

    Args:
        df (pandas.DataFrame): The DataFrame containing surgery records.

    Returns:
        None: The function does not return any value directly. It performs calculations
              and updates the input DataFrame with surgery duration in minutes.
    """
    # Ensure that 'cut time' and 'closing time' columns are of string type
    df['cut time'] = df['cut time'].astype(str)
    df['closing time'] = df['closing time'].astype(str)

    # Convert relevant columns to datetime objects with proper format
    df['Actual date of entry into the operating room'] = pd.to_datetime(df['Actual date of entry into the operating room'] + ' ' + df['cut time'])
    df['Date of end of surgery (exiting the operating room)'] = pd.to_datetime(df['Date of end of surgery (exiting the operating room)'] + ' ' + df['closing time'])

    # Calculate the time difference between entry and ending in minutes
    df['surgery_duration_minutes'] = (df['Date of end of surgery (exiting the operating room)'] - df['Actual date of entry into the operating room']).dt.total_seconds() / 60

    # Replace negative durations with their absolute values (already handled)
    df['surgery_duration_minutes'] = df['surgery_duration_minutes'].abs()

```

```python
# List of DataFrames
dfs = [df_2019, df_2020, df_2021]

# Loop through each DataFrame in the list and call the time_of_surgery function
for i, df in enumerate(dfs, 1):
    print(f"Processing DataFrame df_{i}")
    time_of_surgery(df)
    print(f"DataFrame df_{i} processing completed\n")
```

#time of admission

creating a new variable that contain the admission time in days

```python
import pandas as pd

def hours_admitted_after_surgery(df):
    """
    Calculate the number of hours the patient was admitted after surgery for a given DataFrame.

    This function takes a DataFrame as input, calculates the number of hours the patient was
    admitted after surgery for each surgery record in the DataFrame, and updates the DataFrame
    in place.

    Args:
        df (pandas.DataFrame): The DataFrame containing surgery records.

    Returns:
        None: The function does not return any value directly. It performs calculations
              and updates the input DataFrame with the calculated number of hours.
    """
    # Specify the relevant columns
    discharge_date_column = 'Date of discharge from hospital after surgery'
    discharge_time_column = 'Discharge time after surgery'

    # Ensure that 'Discharge time after surgery' is of timedelta data type
    df[discharge_time_column] = pd.to_timedelta(df[discharge_time_column])

    # Drop rows with null values in the specified columns
    df.dropna(subset=[discharge_date_column, discharge_time_column, 'Date of end of surgery (exiting the operating room)'], inplace=True)

    # Convert discharge date and time columns to datetime objects with proper format
    df[discharge_date_column] = pd.to_datetime(df[discharge_date_column])

    # Combine discharge date and time columns to create datetime for discharge
    df['discharge_datetime'] = df[discharge_date_column] + df[discharge_time_column]

    # Calculate the number of hours the patient was admitted after surgery
    df['hours_admitted_after_surgery'] = ((df['discharge_datetime'] - df['Date of end of surgery (exiting the operating room)']).dt.total_seconds() / 3600).round()   # Convert seconds to hours

    # Replace negative durations with their absolute values (already handled)
    df['hours_admitted_after_surgery'] = df['hours_admitted_after_surgery'].abs()

# Example usage:
# hours_admitted_after_surgery(your_dataframe)
```

```python
# List of DataFrames
dfs = [df_2019, df_2020, df_2021]

# Loop through each DataFrame in the list and call the hours_admitted_after_surgery function
for i, df in enumerate(dfs, 1):
    print(f"Processing DataFrame df_{i}")
    hours_admitted_after_surgery(df)
    print(f"DataFrame df_{i} processing completed\n")
```

#null values on *surgery_duration_minutes* and on *hours_admitted_after_surgery* columns

here we will deal with null values and replace it with the avg for the same surgery

```python
import pandas as pd

# Assuming dfs is a list of DataFrames
for i, df in enumerate(dfs, 1):
    print(f"Processing DataFrame df_{i}")

    # Calculate the average values for each activity
    avg_activity_values = df.groupby('activity')[['surgery_duration_minutes', 'hours_admitted_after_surgery']].transform('mean')

    # Replace null values with the corresponding average values based on 'activity'
    df['surgery_duration_minutes'].fillna(avg_activity_values['surgery_duration_minutes'], inplace=True)
    df['hours_admitted_after_surgery'].fillna(avg_activity_values['hours_admitted_after_surgery'], inplace=True)

    # Verify if there are any remaining null values
    null_values_count = df[['surgery_duration_minutes', 'hours_admitted_after_surgery']].isnull().sum().sum()
    print(f"Number of remaining null values in DataFrame df_{i}: {null_values_count}")

    print(f"DataFrame df_{i} processing completed\n")
```

```python
df_names = ['2019', '2020', '2021']

for df_name, df in zip(df_names, dfs):

    # Save the modified DataFrame as a new CSV file
    output_filename = f"df_{df_name}_modified.csv"
    df.to_csv(output_filename, index=False)

    print(f"DataFrame df_{df_name} saved as {output_filename}\n")
```
